```python
df4['Other_Factors']=df4['Other_Factors'].astype(float)
df4['Num_Injured']=df4['Num_Injured'].astype(float)
df4['DAD']=df4['DAD'].astype(float)
```

```python
df4.info()
```

```python
df4
```

```python
df5=df4.rename(columns={'Quarter': 'Trimestral del año'})
df6=df5.rename(columns={'State': 'Estado'})
df7=df6.rename(columns={'Total_Crashes': 'Total de accidentes'})
df8=df7.rename(columns={'Num_Injured': 'Numero de heridos'})
df9=df8.rename(columns={'Num_Killed': 'Numero de muertos'})
df10=df9.rename(columns={'Total_Vehicles_Involved': 'Total de autos involucrados'})
df11=df10.rename(columns={'SPV': 'Accidentes por exceso de velocidad'})
df12=df11.rename(columns={'DAD': 'Accidentes por influencia de alcohol/drogas'})
df13=df12.rename(columns={'PWR': 'Accidentes por malas condiciones climaticas'})
df14=df13.rename(columns={'FTQ': 'Accidentes por fatiga del conductor'})
df15=df14.rename(columns={'Other_Factors': 'Otros factores'})
df15
```

```python
df15.to_csv("Base_Limpia_Accidentes_Automovilisticos_En_Nigeria.csv", index=False)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=df15[['Total de accidentes']]
y=df15['Total de autos involucrados']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
correlacion_pearson=df15['Total de accidentes'].corr(df15['Total de autos involucrados'], method='pearson')
print(f"coeficiente de correlacion de pearson: {correlacion_pearson}")
```

```python
modelo=LinearRegression()
modelo.fit(x_train, y_train)
y_pred=modelo.predict(x_test)
plt.scatter(x_test, y_test, color='pink', label='datos reales')
plt.plot(x_test, y_pred, color='gray', label='predicciones')
plt.title('regresion lineal')
plt.xlabel('total de accidentes')
plt.ylabel('total de autos involucrados')
plt.legend()
plt.show()
```

```python
plt.scatter(x_test, y_test, color='pink', label='datos reales')
plt.plot(x_test, y_pred, color='gray', label='predicciones')
plt.title('regresion lineal')
plt.xlabel('total de accidentes')
plt.ylabel('autos involucrados')
plt.legend()
x_nuevo=[[500]]
y_nuevo=modelo.predict(x_nuevo)
plt.scatter(x_nuevo, y_nuevo, color='red', label=f'prediccion para {x_nuevo[0][0]} accidentes')
plt.legend()
plt.show()
print(f'la prediccion para {x_nuevo[0][0]} total de accidentes es: {y_nuevo[0]}')
```
