```python
df4.info()
```

```python
df4['Other_Factors']=df4['Other_Factors'].astype(float)
df4['Num_Injured']=df4['Num_Injured'].astype(float)
df4['DAD']=df4['DAD'].astype(float)
```

```python
df4.info()
```

```python
df4
```

```python
df5=df4.rename(columns={'Quarter': 'Trimestral del año'})
df6=df5.rename(columns={'State': 'Estado'})
df7=df6.rename(columns={'Total_Crashes': 'Total de accidentes'})
df8=df7.rename(columns={'Num_Injured': 'Numero de heridos'})
df9=df8.rename(columns={'Num_Killed': 'Numero de muertos'})
df10=df9.rename(columns={'Total_Vehicles_Involved': 'Total de autos involucrados'})
df11=df10.rename(columns={'SPV': 'Accidentes por exceso de velocidad'})
df12=df11.rename(columns={'DAD': 'Accidentes por influencia de alcohol/drogas'})
df13=df12.rename(columns={'PWR': 'Accidentes por malas condiciones climaticas'})
df14=df13.rename(columns={'FTQ': 'Accidentes por fatiga del conductor'})
df15=df14.rename(columns={'Other_Factors': 'Otros factores'})
df15
```

```python
df15.to_csv("Base_Limpia_Accidentes_Automovilisticos_En_Nigeria.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.figure(figsize=(15,6))
plt.hist(df15['Trimestral del año'], bins=30, color='pink', edgecolor='red');
plt.show()
```

```python
gender_counts=df15['Estado'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%3.1f%%', startangle=90, colors=['skyblue', 'pink', 'red'])
plt.title('Distribucion por genero')
plt.show()
```

```python
df_numeric=df15[['Total de accidentes', 'Accidentes por exceso de velocidad', 'Accidentes por influencia de alcohol/drogas', 'Accidentes por malas condiciones climaticas', 'Accidentes por fatiga del conductor', 'Otros factores']]
correlation_matrix=df_numeric.corr()
correlation_matrix
```

```python
plt.figure(figsize=(5,5))

sns.heatmap(correlation_matrix, annot=True, cmap='pink')
plt.show()
```
